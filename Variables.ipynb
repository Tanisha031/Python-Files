```python
#Indexing with [], [:]
var = 'Hello Tanisha!'

print(var)       # Prints complete string
print(var[0])    # Prints first character of the string
print(var[2:5])  # Prints characters starting from 3rd to 5th
print(var[2:])   # Prints string starting from 3rd character
print(var * 2)   # Prints string two times
print(var + "Baby!") # Prints concatenated string
print(var[-1])
```

```python
#Python Data Type Conversion
print("Conversion to integer data type")
a = int(1)     # a will be 1
b = int(2.2)   # b will be 2
c = int("35")   # c will be 3

print (a)
print (b)
print (c)

print("Conversion to floating point number")
q = float(1)     # q will be 1.0
r = float(2.2)   # r will be 2.2
s = float("35") # s will be 3.3

print (q)
print (r)
print (s)

print("Conversion to string")
x = str(1)     # x will be "1"
y = str(2.2)   # y will be "2.2"
z = str("35") # z will be "3.3"

print (x)
print (y)
print (z)

print("Conversion to boolean")
u = bool(1)
v = bool(2.2)
w = bool("3.3")

print (u)
print (v)
print (w)
```
