```python
#Binary String to Integer
#The string should be made up of 1 and 0 only, and the base should be 2.
a = int("110011",2)                           #a = 51
a                                             #Because: 1×2⁵ + 1×2⁴ + 0×2³ + 0×2² + 1×2¹ + 1×2⁰
                                              #    =    32 + 16 + 0 + 0 + 2 + 1 = 51
```

```python
#Hexa-Decimal String to Integer
a = int("2A9", 16)                              #2A9 in hexadecimal = #2 × 16² + A × 16¹ + 9 × 16⁰
a                                                                     #2 × 256 = 512
                                                                      #A = 10 → 10 × 16 = 160
                                                                      #9 × 1 = 9
                                                                      #Total = 512 + 160 + 9 = 681 ✅
```
