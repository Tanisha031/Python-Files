```python
#python Literals orConstants
#......integer literals....(decimal,octal and hexa-decimal)
```

```python
#decimal literal
x = 5
y= -1
z = 0
z1 = 25
print(x,y,z,z1)     #we can also print them separately.
```

```python
# Octal Literal
x = 0O34
y = 0o35
print(x,y)
```

```python
#Hexadecimal Literal
x = 0X34
y = 0x35
z = 0x1C
z1 = 0xC1
print(x,y,z,z1)
```

```python
#example of octal and Hexa-decimal
x = 0O34; y = 0xC1
print("0O34 in octal is ", x, type(x),'&',  "0xC1 in hexa-decimal is" , y, type(y))
```

```python
#.....Float Literal.....
x = 22.24
y=0.12
z = -34.14
z1 = 3.5e3
z2 = -87.7e100
z3 = 2.3e-4
z4 = -5.6e-4
print(x,y,z,z1,z2,z3,z4)
```

```python
print(type(x))
```

```python
#example float in scientific notation
x = 3.5e3
y = 13.2E4
z = -87.7e100
print("3.5e3 in float is", x, type(x),'&', "13.2E4 in float is", y, type(y),'&', "-87.7e100 in float is",type(z))
```

```python
#......Complex Literal......

x= 3 + 5j
y = 10 + 2j
z = 2.5 + 4.1j
print("3+5j in real-imaginary is",x, type(x), '&', "10+2j in real-imaginary is", y, type(y), '&',  "2.5+4.1j in real-imaginary is", z, type(z))
```

```python
#.....String Literal...... ('H', ""H"", """H""", '''H''')
var1 = 'Hello'
var2 = "Hello"
print(" 'Hello' in single quotes is", var1, type(var1), '&', ' "Hello" in double quotes is', var2, type(var2))
```

```python
var3 = """Hello"""
var4 = '''Hello'''
print(' """Hello"""  in double-triple quotes is', var3, type(var3) , '&',  "'''Hello''' in single-triple quotes is", var4, type(var4))
```

```python
#exampleString Literal With Double Quoted Inside String.

var1 = 'my name is "Tanisha" and i am 22 yr old'
print(var1)
```

```python
#......List Literal.......
var1 = [23, 'mingyu', 2.33, True]
print(var1, type(var1))
```

```python
#......Tuple Literal....
var2 = (23, 'woonyong', 4.34, True)
print(var2, type(var2))
```

```python
#tuple type literal without paretheses
var3 = 24, 'joshua', 2.44, False
print(var3, type(var3))
```

```python
#......Dictionary Literals.....
Capitals = {'Nepal': 'Kathmandu', 'Italy': 'Rome', 'England': 'London'}
Numbers = {1: 'one', 2: 'two', 3: 'three'}
Points = {'p1':(10,10), 'p2':(20,20), 'p3':(30,30)}
print(Capitals, type(Capitals), '&', Points, type(Points), '&', Numbers, type(Numbers))
```
